```python
f, ax = plt.subplots(figsize=(8, 2))

ax.set(xlim=(0,12))
ax.set(ylabel=None)
rank = b03['logq_round']

sns.barplot(data = b03, x="-log(q)", y="Description", 
            label="something", 
            orient = "h",
            palette = np.array(pal[::-1])[rank])

ax.set(ylabel=None)
plt.savefig('b03.svg', bbox_inches = 'tight')
```
